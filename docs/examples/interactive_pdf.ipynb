```python
import numpy as np
import plotly.graph_objects as go
import ipywidgets as widgets
from aub_htp import generate_alpha_stable_pdf

# x values
x_vals = np.linspace(-10, 10, 100)

from IPython.display import display

# Create output widget
output = widgets.Output()

# Function to update figure in output
def update_plot_out(alpha, beta, gamma, delta):
    with output:
        output.clear_output(wait=True)  # Clear previous figure
        y = generate_alpha_stable_pdf(x_vals, alpha, beta, gamma, delta)
        fig = go.Figure()
        fig.add_trace(go.Scatter(x=x_vals, y=y, mode='lines', name='Stable PDF'))
        fig.update_layout(
            title=f"Stable PDF (α={alpha}, β={beta}, γ={gamma}, δ={delta})",
            xaxis_title="x",
            yaxis_title="PDF",
            hovermode="x unified"
        )
        fig.show()
alpha_slider = widgets.FloatSlider(value=1.8, min=0.1, max=2, step=0.01, description='α') 
beta_slider = widgets.FloatSlider(value=0, min=-1, max=1, step=0.01, description='β') 
gamma_slider = widgets.FloatSlider(value=1, min=1, max=10, step=0.1, description='γ') 
delta_slider = widgets.FloatSlider(value=0, min=-10, max=10, step=0.1, description='δ')

# Link sliders to output
interactive_plot = widgets.interactive_output(update_plot_out, {
    'alpha': alpha_slider, 
    'beta': beta_slider,
    'gamma': gamma_slider,
    'delta': delta_slider
})
```

```python
# Display sliders and plot
display(widgets.VBox([alpha_slider, beta_slider, gamma_slider, delta_slider, output]))
```
