```python
import matplotlib.pyplot as plt
import numpy as np
from aub_htp.random.spectral_measure_sampler import IsotropicSampler, EllipticSampler, MixedSampler
from aub_htp.random.alpha_stable_sampler import sample_alpha_stable_vector

sigma = [[10, 0],
         [0, 10]]
sampler = EllipticSampler(2, sigma, total_mass = 1)
x = np.asarray([sample_alpha_stable_vector(100, 1.2, sampler) for _ in range(20000)])

plt.scatter(x.T[0], x.T[1])
plt.xlim(-100, 100)
plt.ylim(-100, 100)
# Add title and labels (optional)
plt.title("Simple Scatter Plot")
plt.xlabel("X Value")
plt.ylabel("Y Value")
plt
```
